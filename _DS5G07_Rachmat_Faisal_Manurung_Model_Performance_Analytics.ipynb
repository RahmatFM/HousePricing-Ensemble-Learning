```python
num = ['LotFrontage','LotArea','BsmtFinSF1','TotalBsmtSF','GrLivArea']

for cols in num:
    Q1 = df_merged[cols].quantile(0.25)
    Q3 = df_merged[cols].quantile(0.75)
    IQR=Q3-Q1
    lr= Q1-(1.5 * IQR)
    ur= Q3+(1.5 * IQR)
    df_merged[cols] = df_merged[cols].mask(df_merged[cols]<lr, lr, )
    df_merged[cols] = df_merged[cols].mask(df_merged[cols]>ur, ur, )
```

# Feature Enginnering

> Berdasarkan fitur yang kita miliki saat ini, fitur tambahan pertama yang bisa kita tambahkan adalah TotalLot, yang menjumlahkan LotFrontage dan LotArea untuk mengidentifikasi total luas tanah yang tersedia sebagai kavling. Kita juga bisa menghitung jumlah total luas permukaan rumah, TotalSF dengan menambahkan luas dari basement dan lantai 2. TotalBath juga dapat digunakan untuk mengetahui jumlah kamar mandi yang ada di rumah tersebut. Kita juga bisa menambahkan semua jenis beranda yang berbeda di sekitar rumah dan menggeneralisasikannya ke dalam total luas beranda, TotalPorch.

```python
df_merged['TotalLot'] = df_merged['LotFrontage'] + df_merged['LotArea']
df_merged['TotalBsmtFin'] = df_merged['BsmtFinSF1'] + df_merged['BsmtFinSF2']
df_merged['TotalSF'] = df_merged['TotalBsmtSF'] + df_merged['2ndFlrSF']
df_merged['TotalBath'] = df_merged['FullBath'] + df_merged['HalfBath']
df_merged['TotalPorch'] = df_merged['OpenPorchSF'] + df_merged['EnclosedPorch'] + df_merged['ScreenPorch']
df_merged["saleyr"] = df_merged['YrSold'] - df_merged['YearBuilt']
```

> kita akan melakukan feature enginnering untuk beberapa kolom menggunakan binary yang dapat di artikan **ada(1)/tidak ada(0)**

```python
colum = ['MasVnrArea','WoodDeckSF', 'PoolArea']

for col in colum:
    col_name = col+'_bin'
    df_merged[col_name] = df_merged[col].apply(lambda x: 1 if x > 0 else 0)
```

## Feature Encoding

```python
ordinal_map = {'Ex': 5,'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1, 'NA':0}
fintype_map = {'GLQ': 6,'ALQ': 5,'BLQ': 4,'Rec': 3,'LwQ': 2,'Unf': 1, 'NA': 0}
expose_map = {'Gd': 4, 'Av': 3, 'Mn': 2, 'No': 1, 'NA': 0}
fence_map = {'GdPrv': 4,'MnPrv': 3,'GdWo': 2, 'MnWw': 1,'NA': 0}
```

```python
ord_col = ['ExterQual','ExterCond','BsmtQual', 'BsmtCond','HeatingQC','KitchenQual','GarageQual','GarageCond', 'FireplaceQu']
for col in ord_col:
    df_merged[col] = df_merged[col].map(ordinal_map)

fin_col = ['BsmtFinType1','BsmtFinType2']
for col in fin_col:
    df_merged[col] = df_merged[col].map(fintype_map)

df_merged['BsmtExposure'] = df_merged['BsmtExposure'].map(expose_map)
df_merged['Fence'] = df_merged['Fence'].map(fence_map)
```
